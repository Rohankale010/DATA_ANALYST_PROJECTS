```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth,association_rules
import warnings
warnings.filterwarnings(action='ignore')
```

```python
filepath='D:\Projects\DATA_ANALYST_PROJECTS\Customer_Segmentation_&_Market_Basket_Analysis\data\Market_Basket_Processed.csv'
basket=pd.read_csv(filepath,index_col=0)
```

```python
basket.head()
```

```python
# Apply Fp-Growth Algorithm
frequent_itemset=fpgrowth(basket,min_support=0.01,use_colnames=True)
```

```python
# Generate Association Rules
rules=association_rules(frequent_itemset,metric='lift',min_threshold=1.0)
```

```python
# Filter Important Rules (Strong Rules with high confidence & lift)
rules=rules[(rules['confidence']>=0.2)&(rules['lift']>=1.2)]
```

```python
rules=rules.sort_values(by=['lift','confidence'],ascending=False)
```

```python
rules.to_csv('D:\Projects\DATA_ANALYST_PROJECTS\Customer_Segmentation_&_Market_Basket_Analysis\data/Market_Basket_Rules.csv',index=False)
```
