```python
import pandas as pd
```

```python
filepath='D:\Projects\DATA_ANALYST_PROJECTS\Customer_Segmentation_&_Market_Basket_Analysis\data\Market_Basket_Cleaned.csv'
df=pd.read_csv(filepath)
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'],format="%Y-%m-%d %H:%M:%S")
```

```python
basket=df.pivot_table(index='BillNo',columns='Itemname',values='Quantity',aggfunc='sum')
```

```python
basket
```

```python
basket=basket.applymap(lambda x:1 if x>0 else 0)
```

```python
item_counts=basket.sum(axis=0)
frequent_items=item_counts[item_counts>=5].index
basket=basket[frequent_items]
```

```python
basket
```

```python
filepath='D:\Projects\DATA_ANALYST_PROJECTS\Customer_Segmentation_&_Market_Basket_Analysis\data/Market_Basket_Processed.csv'
basket.to_csv(filepath,index=True)
```
