```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
filepath='D:\Projects\DATA_ANALYST_PROJECTS\Customer_Segmentation_&_Market_Basket_Analysis\data\Mall_Customers_processed.csv'
x_scaled=pd.read_csv(filepath)
```

```python
wcss=[]
K_range=range(1,11)

for k in K_range:
    kmeans=KMeans(n_clusters=k,init='k-means++',random_state=42)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
knee=KneeLocator(K_range,wcss,curve='convex',direction='decreasing')
```

```python
plt.figure(figsize=(8,5))
plt.plot(K_range,wcss,marker='o', linestyle='--')
plt.axvline(x=knee.knee,color='red',linestyle='--')
plt.title('Elbow Method with KneeLocator')
plt.xlabel('Number of Cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
print(f'Optimal Number of Cluster:{knee.knee}')
```
