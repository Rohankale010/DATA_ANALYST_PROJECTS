```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the CSV file
df=pd.read_csv('spotify-2023.csv',encoding='cp1252')
# Displaying first few rows in the dataset
df.head()
```

```python
# Dimension of dataset
print(f'Dimension of Dataset :\n{df.shape}')
```

```python
# Displaying the Information of Dataset
df.info()
```

```python
# Checking Null Values 
df.isna().sum()
```

```python
# Summary Statistics
df.describe().transpose()
```

```python
# Dropping 574 row because of it error value
df=df.drop(index=574)
# Dropping missing value column we dont want to filling inappropriate value in it
df=df.drop(labels=['in_shazam_charts','key'],axis=1)
```

```python
df['in_deezer_playlists']=df['in_deezer_playlists'].str.replace(',','')
df['in_deezer_playlists']=pd.to_numeric(df['in_deezer_playlists'])
df['streams']=pd.to_numeric(df['streams'])
```

```python
df['released_date']=pd.to_datetime(df['released_day'].astype('str')+"-"+df['released_month'].astype('str')+"-"+df['released_year'].astype('str'))
df['released_date']
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),cmap='inferno',linewidths=1,linecolor='black',annot=True)
plt.show()
```

```python
# Top 10 songs on spotify
top_10_songs=df.sort_values(by='streams',ascending=False).head(10)
```

```python
# Visualizing top 10 most stream song on spotify
plt.figure(figsize=(20,10))
sns.barplot(x='track_name',y='streams',data=top_10_songs)
plt.title('Top 10 Most Streamed Song on Spotify')
plt.xlabel('Track Name')
plt.xticks(rotation=65)
plt.ylabel('Streams')
plt.show()
```

```python
# Top 10 most streams songs of 2023
top_10_2023=df[df['released_year']==2023].sort_values(by='streams',ascending=False).head(10)
```

```python
# Visualizing top 10 song of 2023
plt.figure(figsize=(20,10))
sns.barplot(x='track_name',y='streams',data=top_10_2023)
plt.title('Top 10 Most Streamed Song on Spotify in 2023')
plt.xlabel('Track Name')
plt.xticks(rotation=65)
plt.ylabel('Streams')
plt.show()
```

```python
# Group the data by artist name and sum the streams for each artist
artist_streams=df.groupby('artist(s)_name')['streams'].sum().reset_index()

# Sort the DataFrame by the total streams in descending order
top_10_artist=artist_streams.sort_values(by='streams',ascending=False).head(10)

# Display the top 10 artists
top_10_artist
```

```python
# Visualizing top 10 most streams artist on apotify
plt.figure(figsize=(20,10))
sns.barplot(x='artist(s)_name',y='streams',data=top_10_artist)
plt.title('Top 10 Most Streamed Artist on Spotify')
plt.xlabel('Artist Name')
plt.xticks(rotation=65)
plt.ylabel('Streams')
plt.show()
```

```python
playlist_song=df.groupby('track_name')['in_spotify_playlists'].max().reset_index()
top_10_added_song = playlist_song.sort_values(by='in_spotify_playlists',ascending=False).head(10)
top_10_added_song
```

```python
plt.figure(figsize=(20,10))
sns.barplot(y='track_name',x='in_spotify_playlists',data=top_10_added_song,orient='h')
plt.title('Most Added Song on Spotify Playlist')
plt.ylabel('Track Name')
plt.xlabel('Count of Song Added')
plt.show()
```

```python
apple_playlist=df.groupby('track_name')['in_apple_playlists'].max().reset_index()
top_10_apple=apple_playlist.sort_values(by='in_apple_playlists',ascending=False).head(10)
top_10_apple
```

```python
plt.figure(figsize=(20,10))
sns.barplot(y='track_name',x='in_apple_playlists',data=top_10_apple,orient='h')
plt.title('Most Added Song In Apple Playlist')
plt.xlabel('Count of Song Added')
plt.ylabel('Track Name')
plt.show()
```
