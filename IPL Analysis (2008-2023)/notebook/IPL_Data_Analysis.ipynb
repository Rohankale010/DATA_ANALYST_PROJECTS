```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
deliveries_df=pd.read_csv('deliveries.csv')
deliveries_df.head()
```

```python
deliveries_df.shape
```

```python
matches_df=pd.read_csv('matches.csv')
matches_df.head()
```

```python
matches_df.shape
```

```python
ipl_df=pd.merge(deliveries_df,matches_df,left_on='match_id',right_on='id',how='outer')
ipl_df.head()
```

```python
ipl_df.isna().sum()
```

```python
pd.set_option('display.max_columns',None)
```

```python
ipl_df[ipl_df['player_dismissed'].isnull()].head()#.unique()
```

```python

# Count the runs scored by each batsman
runs_count = ipl_df.groupby('batter')['batsman_runs'].sum().reset_index()

# Display the total runs scored by each batsman
most_run=runs_count.sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
print(most_run.head(10))
```

```python
# Visualizing Top 5 Run Scorer
plt.figure(figsize=(15,6))
sns.barplot(data=most_run,x=most_run['batter'].head(10),y=most_run['batsman_runs'].head(10))
plt.title('Most Run Scorer Batsman in IPL')
plt.xlabel('Name')
plt.ylabel('Runs')
plt.show()
```

```python
# Grouping runs by season for a specific player
player_runs = ipl_df[ipl_df['batter'] == 'V Kohli'].groupby('season')['batsman_runs'].sum().reset_index()

# Plotting
plt.figure(figsize=(12,6))
sns.lineplot(data=player_runs, x='season', y='batsman_runs', marker='o')
plt.title('Runs Scored by Virat Kohli Across Seasons')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
most_wicket=ipl_df[ipl_df['is_wicket']==1]['bowler'].value_counts().reset_index(name='Wicket')
most_wicket
```

```python
# Visualize Top 10 Most wicket taking Bowler in IPL
plt.figure(figsize=(15,6))
sns.barplot(data=most_wicket,x=most_wicket['index'].head(10),y=most_wicket['Wicket'].head(10),palette='magma')
plt.title('Most Wicket-taking Bowler in IPL')
plt.xlabel('Name')
plt.ylabel('Wickets')
plt.show()
```

```python
# Grouping runs by season for a specific player
player_runs = ipl_df[ipl_df['bowler'] == 'DJ Bravo'].groupby('season')['is_wicket'].sum().reset_index()

# Plotting
plt.figure(figsize=(12,6))
sns.lineplot(data=player_runs, x='season', y='is_wicket', marker='o',color='purple')
plt.title('Wicket Taken by Dj Bravo Across Seasons')
plt.xlabel('Season')
plt.ylabel('Wicket')
plt.show()
```

```python
most_boundaries=ipl_df[ipl_df['batsman_runs'].isin([4,6])]['batter'].value_counts().reset_index()
most_boundaries.columns=['batter','boundaries']
most_boundaries.head(10)
```

```python
# Calculate number of 4s and 6s for each batsman
boundary_counts = ipl_df.groupby('batter')['batsman_runs'].value_counts().unstack(fill_value=0) 
#unstack is use the multi-index for creating column by choosimg one of index

# Only keep columns for 4s and 6s
boundary_counts = boundary_counts[[4, 6]].rename(columns={4: 'Fours', 6: 'Sixes'})

# Calculate total boundaries
boundary_counts['Total_Boundaries'] = boundary_counts['Fours'] + boundary_counts['Sixes']

# Select top 5 batsmen based on total boundaries
top_10_batsmen = boundary_counts.sort_values(by='Total_Boundaries',ascending=False).head(10)

# Plotting
top_10_batsmen[['Fours', 'Sixes']].plot(kind='bar', stacked=True,figsize=(15,5))
plt.title('Top 10 Batsmen by Boundaries')
plt.xlabel('Batsman')
plt.ylabel('Number of Boundaries')
plt.xticks(rotation=0)
plt.legend(title='Type of Boundary')
plt.show()
```

```python
#team=['Mumbai Indians','Chennai Super Kings','Royal Challengers Bangalore','Sunrisers Hyderabad','Delhi Capitals','Delhi Daredevils','Kolkata Knight Riders','Rajasthan Royals','Gujarat Lions','Gujarat Titans','Kings XI Punjab','Lucknow Super Giants','Deccan Chargers','Rising Pune Supergiant']
# Filter out the Non League Matches
qualify=ipl_df[ipl_df['match_type']!='League']

#Groupby to Count the versus match of team according to season
qualifiers=qualify.groupby(['season','team1','team2']).size().reset_index() # size is used to count the occurance

# Count the number of occurance of each team in team1
team1_counts=qualifiers['team1'].value_counts().reset_index()
# Count the number of occurance of each team in team2
team2_counts=qualifiers['team2'].value_counts().reset_index()

# Merge the both team counts
team=pd.merge(team1_counts,team2_counts,left_on=team1_counts['index'],right_on=team2_counts['index'],how='outer')
team.fillna(0,inplace=True)

# Sum the number of occurance of each team to get total 
team['Qualified']=team['team1'].astype('int')+team['team2'].astype('int')
team.rename(columns={'key_0':'team_name'},inplace=True)
team=team[['team_name','Qualified']].sort_values(by='Qualified',ascending=False)
team
```

```python
# Visualizing the Most Qualifying team in IPL
plt.figure(figsize=(20,8))
sns.barplot(data=team,y=team['team_name'],x=team['Qualified'],orient='h',palette='Spectral')
plt.title('Most Qualifying Teams in IPL')
plt.xlabel('Number of Times')
plt.ylabel('Teams')
plt.show()
```

```python
ipl_2023=ipl_df[ipl_df['season']=='2023']
ipl_2023.head()
```

```python
runs_2023=ipl_2023.groupby('batter')['batsman_runs'].sum().reset_index(name='runs')
mostruns_2023=runs_2023.sort_values(by='runs',ascending=False)
mostruns_2023.head()
```

```python
# Visualizing top 5 Most run scorer in 2023
plt.figure(figsize=(10,6))
sns.barplot(data=mostruns_2023,x=mostruns_2023['batter'].head(),y=mostruns_2023['runs'].head())#,color='orange')
plt.title('Most Runs Score in IPL 2023')
plt.xlabel('Batsman Name')
plt.ylabel('Runs')
plt.show()
```

```python
mostwicket_2023=ipl_2023[ipl_2023['is_wicket']==1]['bowler'].value_counts().reset_index(name='wickets')
mostwicket_2023.head()
```

```python
 # Visualizing top 5 Most run scorer in 2023
plt.figure(figsize=(10,6))
sns.barplot(data=mostwicket_2023,x=mostwicket_2023['index'].head(),y=mostwicket_2023['wickets'].head(),palette='viridis')
plt.title('Most Wicket-taking Bowler in IPL 2023')
plt.xlabel('Bowler Name')
plt.ylabel('Wickets')
plt.show()
```

```python
trophy=ipl_df[ipl_df['match_type']=='Final'][['winner','season']].value_counts().reset_index()
trophy=trophy['winner'].value_counts().reset_index(name='trophy')
trophy.rename(columns=({'index':'Team_Name'}),inplace=True)
trophy
```

```python

# Visualizaton for Number of Trophy Wins
plt.pie(x=trophy['trophy'],labels=trophy['Team_Name'],autopct=lambda trophy_count: f"{trophy_count * trophy['trophy'].sum() / 100:.0f}")
# percentage of one part is shown when we put autocpt='%.2f%%'
# trophy_count is percent of one section so the calculation is like that 16x31.25/100=5
plt.title('Number of Trophy Win by a Team')
plt.show()
```
