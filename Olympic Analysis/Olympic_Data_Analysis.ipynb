```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Dataset
athlete=pd.read_csv('athlete_events.csv')

# Display the first few row of dataset
athlete.head()
```

```python
# Load the dataset
region=pd.read_csv('noc_regions.csv')

# Display the first few row of dataset
region.head()
```

```python
# Merge the two dataset
athlete_df=athlete.merge(region,how='left',on='NOC')
athlete_df.head()
```

```python
# Dimension 
athlete_df.shape
```

```python
athlete_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
```

```python
# Display Information
athlete_df.info()
```

```python
athlete_df.describe()
```

```python
# India Details
athlete_df.query('Team=="India"').head()
```

```python
# Japan Details
athlete_df.query('Team=="Japan"')
```

```python
# Top Coutries Participating
top_10_countries=athlete_df['Team'].value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Visualizing 
plt.figure(figsize=(12,7))
sns.barplot(x=top_10_countries.index,y=top_10_countries.values)
plt.title('Overall Participation by Country')
```

```python
plt.figure(figsize=(12,6))
plt.hist(x=athlete_df['Age'],bins=np.arange(10,80,2),color='orange',edgecolor='white')
plt.title('Age Distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.show()
```

```python
# Winter olympic sport
athlete_df[athlete_df['Season']=='Winter'].Sport.unique()
```

```python
# Summer olympic sport
athlete_df[athlete_df['Season']=='Summer'].Sport.unique()
```

```python
gender_counts=athlete_df['Sex'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(10,6))
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

```python
# Total Number of female athletes in each olympics
female_participants=athlete_df[(athlete_df['Sex']=='F')&(athlete_df['Season']=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants
```

```python
women_olympics_summer=athlete_df[(athlete_df['Sex']=='F')&(athlete_df['Season']=='Summer')]
plt.figure(figsize=(15,7))
sns.countplot(data=women_olympics_summer,x='Year',palette='Spectral')
plt.title('Women Participation in Summer')
plt.show()
```

```python
women_olympics_winter=athlete_df[(athlete_df['Sex']=='F')&(athlete_df['Season']=='Winter')]
plt.figure(figsize=(15,7))
sns.countplot(data=women_olympics_winter,x='Year',palette='Spectral')
plt.title('Women Participation in Summer')
plt.show()
```

```python
part_summer=women_olympics_summer.groupby('Year')['Sex'].value_counts()
part_winter=women_olympics_winter.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(15,7))
part_summer.loc[:,'F'].plot()
part_winter.loc[:,'F'].plot()
plt.title('Women Athlete Participation over Time in Different Season')
plt.show()
```
