```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df=pd.read_csv('D:\Projects\DATA_ANALYST_PROJECTS\Financial_Fraud_Detection\data\creditcard.csv')
```

```python
# Summary Statistics
df.describe()
```

```python
# Count of Fraudalent and Non Fraudalent transaction
fraud_counts=df['Class'].value_counts()
fraud_counts
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=fraud_counts.index,y=fraud_counts.values,palette=['blue','red'])
plt.xticks(ticks=[0,1],labels=['Legit(0)','Fraud(1)'])
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
print(f'Legimate Transactions : {fraud_counts[0]}')
print(f'Fraudalent Transactions : {fraud_counts[1]}')
print(f'Fraud Percentage : {(fraud_counts[1]/fraud_counts.sum())*100 :.2f}%')
```

```python
# Checking for scaling Required
plt.figure(figsize=(20,15))
sns.histplot(data=df,bins=50)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=False,linecolor='black',linewidths=1.2)
plt.title('Feature Correlation')
plt.show()
```
